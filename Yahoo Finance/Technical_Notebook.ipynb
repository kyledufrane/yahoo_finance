### Coins

```python
public_client = cbpro.PublicClient()

coins = pd.DataFrame(public_client.get_products())

coins.head()
```

```python
no_yahoo = []

start = datetime(2016,1,1)
end = datetime.today()

for ticker in coins['id']:
    
    tckr = yf.Ticker(ticker)
    
    historical = tckr.history(start=start, 
                              end=end, 
                              interval="1d")
    if len(historical) > 0:
    
        pickle_out = open(f'crypto_history_1d/{ticker}.csv',"wb")

        pickle.dump(historical, pickle_out)

        pickle_out.close()
        
    else:
        
        no_yahoo.append(ticker)
```
